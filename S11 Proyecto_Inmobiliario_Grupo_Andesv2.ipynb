# 📊 Proyecto: Análisis Comercial Inmobiliario

La empresa Andes Capital Real State del sector inmobiliario necesita evaluar su **desempeño comercial** para comprender su crecimiento, rentabilidad y comportamiento de clientes.

Se solicita construir un **dashboard ejecutivo** que permita analizar ventas, clientes y propiedades para apoyar decisiones estratégicas basadas en datos.


# 📂 Descripción de los Datasets

🔹 **hecho_ventas_propiedades**  
Tabla de hechos con las transacciones de venta (precio, cliente, propiedad, canal, fecha).

🔹 **dim_clientes**  
Información y segmentación de clientes.

🔹 **dim_propiedades**  
Características de las propiedades (tipo, tamaño, ubicación, etc).

🔹 **dim_fecha**  
Tabla calendario que deberá ser creada como parte del modelado.

# 🎯 Preguntas que el dashboard debe responder

**Desempeño general**

- ¿Cuál es el **ingreso total** generado por la empresa?
- ¿Cuántas **propiedades se han vendido**?
- ¿Cuál es el **precio promedio de venta**?
- ¿Cuál es la **comisión total** generada por la empresa?

**Análisis comercial**

- ¿Qué **tipo de propiedad genera más ingresos**?
- ¿Qué **segmento de clientes compra más propiedades**?
- ¿Qué canal de venta **genera mayor ingresos**?


**Análisis temporal**

- ¿Cómo evolucionan las **ventas en el tiempo**?
- ¿El negocio está **creciendo año contra año**?
- ¿Cómo va el desempeño **acumulado del año actual (YTD)**?

**Cohortes de clientes**

- ¿Los clientes **vuelven a comprar** después de su primera compra?
- ¿Qué **cohortes de clientes generan más compras/ingresos con el tiempo**?

## 📊 Paso 4: Creación de medidas

En este paso se crean las **medidas necesarias para responder las preguntas de negocio del dashboard**.

Las medidas se implementan como DAX en **Power BI** (o como Calculated Fields en Tableau).

### 4.1 🔹 Medidas base

Se crean las siguientes medidas:

- **Ingreso Total** = SUM(hecho_ventas_propiedades[precio_venta])
- **Cantidad de Ventas** = DISTINCTCOUNT(hecho_ventas_propiedades[id_venta])
- **Ticket Promedio** = DIVIDE([Ingreso Total],[Cantidad de Ventas])
- **Comisión Total** =  SUM(hecho_ventas_propiedades[monto_comision])

Estas métricas permitirán analizar el desempeño general del negocio.

### 4.2 🔹 Medidas con contexto de filtro

Se crean 3 medidas que modifican el contexto de filtro.

En este caso se utilizarán **medidas de participación (%)**, que permiten entender **qué proporción del ingreso total aporta cada categoría**.

Las medidas que deben construirse son:

- **Participación de ingresos por tipo de propiedad**= DIVIDE([Ingreso Total],CALCULATE([Ingreso Total],ALL(hecho_ventas_propiedades[tipo_propiedad])))
- **Participación de ingresos por canal de venta**= DIVIDE([Ingreso Total],CALCULATE([Ingreso Total],ALL(hecho_ventas_propiedades[canal_venta])))
- **Participación de ingresos por segmento de cliente**= DIVIDE([Ingreso Total],CALCULATE([Ingreso Total],ALL(dim_clientes[segmento_comprador])))

Para ello:
- Se modifica el contexto de filtro con las funciones propias de cada herramienta
  - En Power BI, usa `CALCULATE`
  - En Tableau, usa expresiones como `FIXED` (LOD)

Se elimina el filtro de la dimensión correspondiente para obtener el total

En este caso, el denominador represente el total sin el filtro de la categoría, mientras que el numerador sí lo conserva.

Estas medidas se utilizarán principalmente como **tooltip o etiqueta adicional en los gráficos** del dashboard para complementar el análisis.

### 4.3 🔹 Inteligencia de tiempo

Utilizando la tabla **`dim_fecha`** (Power BI) o las funciones de fecha en Tableau, se crean las siguientes medidas:

- **Ventas Year to Date (YTD)** = TOTALYTD([Ingreso Total],dim_fecha[Date])
- **Ventas Month to Date (MTD)** = TOTALMTD([Ingreso Total],dim_fecha[Date])
- **Ventas YTD Año anterior** = CALCULATE([Ventas YTD],SAMEPERIODLASTYEAR(dim_fecha[Date]))
- **Crecimiento Year over Year (YoY)** = DIVIDE([Ventas YTD]-[Ventas YTD Año anterior],[Ventas YTD Año anterior])

Estas métricas permiten analizar la evolución del negocio en el tiempo.


### 4.4 🔹 Medidas para análisis de cohortes

Creamos cálculos que permitan analizar **si los clientes vuelven a comprar después de su primera compra**.

⚠️ Estos cálculos se crean como **columnas calculadas**, no como medidas.  
Todas se crean en la tabla **`hecho_ventas_propiedades`**.

Creamos las siguientes columnas:

- **Primera compra por cliente** → fecha mínima de compra de cada cliente.  
  primera_compra = CALCULATE(MIN(hecho_ventas_propiedades[fecha_venta]),ALLEXCEPT(hecho_ventas_propiedades,hecho_ventas_propiedades[id_cliente]))

- **Mes Cohorte** → mes en que el cliente realizó su **primera compra** (mes de adquisición).  
  cohorte = FORMAT(hecho_ventas_propiedades[primera compra],"YYYY-MM")
  
- **Mes Venta** → mes en que ocurre cada transacción. Este campo permite identificar **cuándo el cliente vuelve a comprar** después de su primera compra.  
  meses_desde_inicio = DATEDIFF(hecho_ventas_propiedades[primera_compra],hecho_ventas_propiedades[fecha_venta],MONTH)

Estos cálculos permitirán construir una **matriz de cohortes** donde:

- **Filas** representen el **mes de cohorte**
- **Columnas** representen el **mes de venta**
- **Valores** muestren **cantidad de clientes en cada mes que vuelven a comprar**. Para eso debemos crear una nueva medidad:  
  n_clientes = DISTINCTCOUNT(hecho_ventas_propiedades[id_cliente])

- Para facilitar el análisis, creamos medida de Retención %:  
  **Retención %** = DIVIDE(DISTINCTCOUNT(hecho_ventas_propiedades[id_cliente]),                      CALCULATE(DISTINCTCOUNT(hecho_ventas_propiedades[id_cliente]),hecho_ventas_propiedades[meses_desde_inicio]=0))

  La Retención % nos muestra para cada cohorte, el porcentaje de clientes que vuelven a comprar cada mes.

## 📝Paso 6:  Resumen Ejecutivo

**Hallazgos clave**

- El ingreso total del período fue de \$6.000 millones generado por 8.500 ventas.
- El tipo de propiedad que genera mayor revenue son las casas con 37,25% por ciento de las ventas.  
- La ciudad con mayor volumen de ventas es Ciudad de México con 53,9% de las ventas.  
- El canal de venta más eficiente en términos de ingresos es corredor con un 72,85% de las ventas.  
- El crecimiento de 11,14% YoY está bien, pero podría ser mejor.  
- El ticket promedio de \$707.353 indica que el crecimiento viene dado por volumen, no por valor premium. Estamos jugando en el segmento masivo (primera vivienda con un 62,93%), que es el motor actual de las ventas.  
- EL mix de productos es bastante equilibrado (casas 37,25%; comercial 32,01% y deptos 30,72%), pero no hay un ganador claro. No estamos explotando una especialización dominante.  
- El principal riesgo es la dependencia de dos ciudades para las ventas (Ciudad de México y Bogotá). Si se frenan, sufre la caja.  
- Los corredores representan el mayor canal de venta, con lo que hay un menor control de la experiencia del cliente, margen presionado (comisiones) y dependencia de terceros.  
- Las cohortes del 2023 son mejores que las del 2024, lo cual indica que no hay una mejora progresiva. Algo se hizo bien el 2023 que después no se llevó a la práctica el 2024.  
- Existe un flujo constante de ventas en sector masivo con alta dependencia de terceros (corredores) y concentrado en dos ciudades, con una recurrencia débil, propia del segmenteo primera vivienda.  
- El negocio se ve estable, pero no se está construyendo un sistema que genero un flujo constante en crecimiento año tras año más agresivo.  

**Métricas principales**
- Ingreso Total:  \$6.012.502.170  
- Cantidad de Ventas: 8.500  
- Ticket Promedio:  \$ 707.353   
- Comisión Total: \$200.627.166   

**Insights accionables**
- El segmento de clientes que compran por primera vez concentra la mayor parte del ingreso con un 62,93%.

- Los clientes adquiridos en las cohortes de 2023-03, 2023-07, 2023-10 y 2024-03 muestran mayor recurrencia:  
  a) **2023-03**: alta desde el inicio (13%-12%); se mantiene fuerte hasta los meses 6-10; tiene picos de 16%; la más sólida y consistente.  
  b) **2023-07**: arranca bajo (4,69%) pero meses 4-9 muy fuertes (14%-12%-11%-10%); buena resistencia en el tiempo.  
  c) **2023-10**: varios meses sobre 10-13%; sin caídas bruscas; muy pareja (perfil cliente fiel, no impulsivo)  
  d) **2024-03**: cohorte muy interesante aunque con pocos datos por ser reciente; arranca con 17,07% y se mantiene alto; alta promesa pero aún inmadura.
- Las ventas muestran un crecimiento de 11,14% YoY.
- Hay un alza de ventas en Q4, con Q3 con el menor nivel de ventas.
  
**Recomendaciones estratégicas**

1.  **Subir el ticket promedio.**
   
   - Upselling (mejor ubicación, extras)
   - Upgrades (estacionamiento, bodegas, áreas verdes)
   - Financiamiento estructurado
   - Meta: +10-15% ticket > impacto inmediato en revenue.
     
2.  **Desarrolar segmento inversionista.**
   
   - Ofertas diseñadas para renta.
   - Reportes de rentabilidad (tipo mini asset management)
   - Programas de recompra/recompra preferente.
   - Este segmento es el que genera recurrencia real.
   - Meta: 24% a 40% de las ventas  

3.   **Reducir dependencia de corredores.**

   - No eliminarlos pero:
   - Fortalecer canal directo con:
         a) Marketing digital.
         b) CRM + seguimiento.
   - Metas: 27% a 40% directo
  
4.  **Diversificación geográfica.**
   
   - Replicar lo que funciona en Ciudad de Mëxico y Bogotá.
   - Sin copiar producto sino la propuesta de valor.

5.  **Optimizar recurrencia.**
   
    - Postventa activa (no desaparecer después de vender).
    - Cross-sell (segunda propiedad)
    - Programas "cliente inversionista"
    - Meta: convertir comprador en inversionista.
